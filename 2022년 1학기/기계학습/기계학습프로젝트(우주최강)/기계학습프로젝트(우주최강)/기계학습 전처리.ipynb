```python
import glob

all_files=glob.glob('wow*.csv')

all_files_data=[]

for file in all_files:
    data_frame=pd.read_csv(file, encoding='cp949')
    all_files_data.append(data_frame)
    
all_files_data_concat=pd.concat(all_files_data, axis=1)

all_files_data_concat.to_csv('전처리 최종.csv', encoding='cp949', index=False)
```
