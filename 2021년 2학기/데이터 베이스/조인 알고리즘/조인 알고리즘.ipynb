```python
import pandas as pd
  
def length(filename): #해당파일의 2번째 행부터 끝 행까지의 개수를 나타내는 함수
    with open(filename,'r')as f:
        rows=f.readlines()[1:]
    return len(rows) 
    
filename1,filename2=input("조인할 파일 2개를 입력하세요: ").split(' ')
        
join_number=input("조인 방법을 선택하세요 (1. Nested loop 2. Block Nested loop):")
join_number=int(join_number)
        
col1,col2=input("조인할 각 컬럼 값들을 입력하세요:").split(' ')
col1=int(col1)
col2=int(col2)

n=input("한 블록 내의 투플 개수를 입력하세요:")    
n=int(n)
        
output_file=input("출력할 파일명을 입력하세요:")
        
l1=length(filename1) #두 파일의 길이 저장
l2=length(filename2)

if l1%n==0: #한 블록내의 투플개수에따라 총 블록의개수와 블록내의 투플수가 달라진다
    num1=l1//n
    res1=0
else:
    num1=(l1//n)+1
    res1=l1%n

if l2%n==0:
    num2=l2//n
    res2=0
else:
    num2=(l2//n)+1
    res2=l2%n          

    def bring_block(filename,row1,row2): #지정 범위내의 투플들을 가져와서 나타내는 함수
        with open(filename,'r')as f:
                rows=f.readlines()[row1:row2]
        for x in range(len(rows)):
            rows[x]=rows[x].split()
        result=pd.DataFrame(rows)   
        return result  

    def Nested_Loop_Join(): 
        output=pd.DataFrame()
        if res1>0: #첫번째파일이 한 블록내의 투플로 나누어 떨어지지 않을때
            for x in range(num1):
                if x==num1-1:
                    for j in range(res1):
                        if res2>0: #두번째파일이 한 블록내의 투플로 나누어 떨어지지 않을때
                            for i in range(num2):
                                if i==num2-1:
                                    input1=bring_block(filename1,(n*x)+1,l1+1) #해당범위의 투플들을 가져온다
                                    input2=bring_block(filename2,(n*i)+1,l2+1)
                                    for k in range(res2):
                                        if input1.loc[j,col1-1]==input2.loc[k,col2-1]: #지정된 열에서의 각각의 행의값이 같으면 조인
                                            sample=pd.concat([input1.loc[j,:],input2.loc[k,:]])
                                            output=pd.concat([output,sample],axis=1)
                                        else:
                                            continue
                                else:
                                    input1=bring_block(filename1,(n*x)+1,l1+1)
                                    input2=bring_block(filename2,(n*i)+1,(n*(i+1))+1)
                                    for k in range(n):
                                        if input1.loc[j,col1-1]==input2.loc[k,col2-1]:
                                            sample=pd.concat([input1.loc[j,:],input2.loc[k,:]])
                                            output=pd.concat([output,sample],axis=1)
                                        else:
                                            continue
                        else:
                            for i in range(num2):
                                input1=bring_block(filename1,(n*x)+1,l1+1)
                                input2=bring_block(filename2,(n*i)+1,(n*(i+1))+1)
                                for k in range(n):
                                    if input1.loc[j,col1-1]==input2.loc[k,col2-1]:
                                        sample=pd.concat([input1.loc[j,:],input2.loc[k,:]])
                                        output=pd.concat([output,sample],axis=1)
                                    else:
                                        continue

                else:
                    for j in range(n):
                        if res2>0: 
                            for i in range(num2):
                                if i==num2-1:
                                    input1=bring_block(filename1,(n*x)+1,(n*(x+1))+1)
                                    input2=bring_block(filename2,(n*i)+1,l2+1)
                                    for k in range(res2):
                                        if input1.loc[j,col1-1]==input2.loc[k,col2-1]:
                                            sample=pd.concat([input1.loc[j,:],input2.loc[k,:]])
                                            output=pd.concat([output,sample],axis=1)
                                        else:
                                            continue
                                else:
                                    input1=bring_block(filename1,(n*x)+1,(n*(x+1))+1)
                                    input2=bring_block(filename2,(n*i)+1,(n*(i+1))+1)
                                    for k in range(n):
                                        if input1.loc[j,col1-1]==input2.loc[k,col2-1]:
                                            sample=pd.concat([input1.loc[j,:],input2.loc[k,:]])
                                            output=pd.concat([output,sample],axis=1)
                                        else:
                                            continue

                        else:
                            for i in range(num2):
                                input1=bring_block(filename1,(n*x)+1,(n*(x+1))+1)
                                input2=bring_block(filename2,(n*i)+1,(n*(i+1))+1)
                                for k in range(n):
                                    if input1.loc[j,col1-1]==input2.loc[k,col2-1]:
                                        sample=pd.concat([input1.loc[j,:],input2.loc[k,:]])
                                        output=pd.concat([output,sample],axis=1)
                                    else:
                                        continue
        else:
            for x in range(num1):
                for j in range(n):
                    if res2>0: 
                        for i in range(num2):
                            if i==num2-1:
                                input1=bring_block(filename1,(n*x)+1,(n*(x+1))+1)
                                input2=bring_block(filename2,(n*i)+1,l2+1)
                                for k in range(res2):
                                    if input1.loc[j,col1-1]==input2.loc[k,col2-1]:
                                        sample=pd.concat([input1.loc[j,:],input2.loc[k,:]])
                                        output=pd.concat([output,sample],axis=1)
                                    else:
                                        continue
                            else:
                                input1=bring_block(filename1,(n*x)+1,(n*(x+1))+1)
                                input2=bring_block(filename2,(n*i)+1,(n*(i+1))+1)
                                for k in range(n):
                                    if input1.loc[j,col1-1]==input2.loc[k,col2-1]:
                                        sample=pd.concat([input1.loc[j,:],input2.loc[k,:]])
                                        output=pd.concat([output,sample],axis=1)
                                    else:
                                        continue

                    else:
                        for i in range(num2):
                            input1=bring_block(filename1,(n*x)+1,(n*(x+1))+1)
                            input2=bring_block(filename2,(n*i)+1,(n*(i+1))+1)
                            for k in range(n):
                                if input1.loc[j,col1-1]==input2.loc[k,col2-1]:
                                    sample=pd.concat([input1.loc[j,:],input2.loc[k,:]])
                                    output=pd.concat([output,sample],axis=1)
                                else:
                                    continue

        return output.transpose()

    def Block_Nested_Loop_Join():
        output=pd.DataFrame()
        if res1>0: 
            for x in range(num1):
                if x==num1-1:
                    for i in range(num2):
                        if i==num2-1:
                            for j in range(res1): #해당 블록내의 값들을 조인하고 다음블록으로 넘어간다
                                if res2>0:
                                    input1=bring_block(filename1,(n*x)+1,l1+1)
                                    input2=bring_block(filename2,(n*i)+1,l2+1)
                                    for k in range(res2):
                                        if input1.loc[j,col1-1]==input2.loc[k,col2-1]:
                                            sample=pd.concat([input1.loc[j,:],input2.loc[k,:]])
                                            output=pd.concat([output,sample],axis=1)
                                        else:
                                            continue
                                else:
                                    input1=bring_block(filename1,(n*x)+1,l1+1)
                                    input2=bring_block(filename2,(n*i)+1,(n*(i+1))+1)
                                    for k in range(n):
                                        if input1.loc[j,col1-1]==input2.loc[k,col2-1]:
                                            sample=pd.concat([input1.loc[j,:],input2.loc[k,:]])
                                            output=pd.concat([output,sample],axis=1)
                                        else:
                                            continue
                        else:
                            for j in range(res1):
                                input1=bring_block(filename1,(n*x)+1,l1+1)
                                input2=bring_block(filename2,(n*i)+1,(n*(i+1))+1)
                                for k in range(n):
                                    if input1.loc[j,col1-1]==input2.loc[k,col2-1]:
                                        sample=pd.concat([input1.loc[j,:],input2.loc[k,:]])
                                        output=pd.concat([output,sample],axis=1)
                                    else:
                                        continue

                else:
                    for i in range(num2):
                        if i==num2-1:
                            for j in range(n):
                                if res2>0:
                                    input1=bring_block(filename1,(n*x)+1,(n*(x+1))+1)
                                    input2=bring_block(filename2,(n*i)+1,l2+1)
                                    for k in range(res2):
                                        if input1.loc[j,col1-1]==input2.loc[k,col2-1]:
                                            sample=pd.concat([input1.loc[j,:],input2.loc[k,:]])
                                            output=pd.concat([output,sample],axis=1)
                                        else:
                                            continue
                                else:
                                    input1=bring_block(filename1,(n*x)+1,(n*(x+1))+1)
                                    input2=bring_block(filename2,(n*i)+1,(n*(i+1))+1)
                                    for k in range(n):
                                        if input1.loc[j,col1-1]==input2.loc[k,col2-1]:
                                            sample=pd.concat([input1.loc[j,:],input2.loc[k,:]])
                                            output=pd.concat([output,sample],axis=1)
                                        else:
                                            continue

                        else:
                            for j in range(n):
                                input1=bring_block(filename1,(n*x)+1,(n*(x+1))+1)
                                input2=bring_block(filename2,(n*i)+1,(n*(i+1))+1)
                                for k in range(n):
                                    if input1.loc[j,col1-1]==input2.loc[k,col2-1]:
                                        sample=pd.concat([input1.loc[j,:],input2.loc[k,:]])
                                        output=pd.concat([output,sample],axis=1)
                                    else:
                                        continue
        else:
            for x in range(num1):
                for i in range(num2):
                    if i==num2-1:
                        for j in range(n):
                            if res2>0:
                                input1=bring_block(filename1,(n*x)+1,(n*(x+1))+1)
                                input2=bring_block(filename2,(n*i)+1,l2+1)
                                for k in range(res2):
                                    if input1.loc[j,col1-1]==input2.loc[k,col2-1]:
                                        sample=pd.concat([input1.loc[j,:],input2.loc[k,:]])
                                        output=pd.concat([output,sample],axis=1)
                                    else:
                                        continue
                            else:
                                input1=bring_block(filename1,(n*x)+1,(n*(x+1))+1)
                                input2=bring_block(filename2,(n*i)+1,(n*(i+1))+1)
                                for k in range(n):
                                    if input1.loc[j,col1-1]==input2.loc[k,col2-1]:
                                        sample=pd.concat([input1.loc[j,:],input2.loc[k,:]])
                                        output=pd.concat([output,sample],axis=1)
                                    else:
                                        continue

                    else:
                        for j in range(n):
                            input1=bring_block(filename1,(n*x)+1,(n*(x+1))+1)
                            input2=bring_block(filename2,(n*i)+1,(n*(i+1))+1)
                            for k in range(n):
                                if input1.loc[j,col1-1]==input2.loc[k,col2-1]:
                                    sample=pd.concat([input1.loc[j,:],input2.loc[k,:]])
                                    output=pd.concat([output,sample],axis=1)
                                else:
                                    continue

        return output.transpose()
    
if join_number==1: 
    output=Nested_Loop_Join()
    count=((l1*num2)+num1)#블록의 접근횟수
else:
    output=Block_Nested_Loop_Join()
    count=((num1)*(num2+1))

output.to_csv(output_file,index=False,sep=' ',header=None) #출력파일작성
row_count=len(output) #출력된 파일의 총 행의개수
print("Result rows:",row_count,"\n블록의 접근횟수 :",count)
```
